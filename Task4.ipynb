```python
# Problem 1: Mobile Recharge Validation System


def process_recharge():
    """
    Handles the user interface and retry logic for mobile recharging.
    """
    # Define valid plan constants
    MIN_RECHARGE = 50
    VALID_PLANS = [199, 299, 399, 599]
    
    print("Mobile Recharge Validation System")
    print("-" * 30)
    
    is_active = True
    
    while is_active:
        user_input = input("Enter recharge amount: ")
        
        # Check if input is a digit to avoid crashes
        if not user_input.isdigit():
            print("Error: Please enter a valid numerical amount.")
            continue
            
        amount = int(user_input)
        
        # Rule 1: Minimum Amount Validation
        if amount < MIN_RECHARGE:
            print(f"Invalid: Amount must be at least ₹{MIN_RECHARGE}.")
            print("Please try again.")
            continue
            
        # Rule 2: Plan Selection Validation
        if amount in VALID_PLANS:
            print(f"Recharge successful for the ₹{amount} plan.")
            is_active = False  # Exit condition
        else:
            print(f"Error: ₹{amount} is not a recognized plan.")
            print(f"Available plans: {VALID_PLANS}")
            print("Please enter a valid plan amount.")

if __name__ == "__main__":
    process_recharge()
```

```python
# Problem 2: Inventory Reorder Alert System

def check_inventory(products):
    """
    Scans the inventory dictionary and alerts the user
    if any product stock is below the threshold.
    """
    THRESHOLD = 15
    
    print("Inventory Status Report")
    print("-" * 30)
    
    # Iterate through the dictionary items (Product Name, Quantity)
    for product, stock in products.items():
        if stock < THRESHOLD:
            status = "REORDER ALERT"
            detail = f"Low stock detected: {stock} units remaining."
        else:
            status = "Stock OK"
            detail = f"Current level: {stock} units."
            
        print(f"Product: {product:12} | Status: {status:15} | {detail}")

# Sample Data: Dictionary of products and their quantities
warehouse_stock = {
    "Laptops": 25,
    "Keyboards": 8,
    "Monitors": 12,
    "Mouse": 40,
    "Webcams": 5
}

# Execute the check
if __name__ == "__main__":
    check_inventory(warehouse_stock)
```

```python
# Problem 3: Student Result Processing System


def process_student_result(marks_list):
    """
    Calculates the average of marks and determines Pass/Fail status.
    """
    if not marks_list:
        return 0, "No data"

    # Rule 1: Calculate total using a loop
    total_score = 0
    for mark in marks_list:
        total_score += mark
    
    # Rule 2: Calculate average
    average = total_score / len(marks_list)
    
    # Rule 3: Determine status based on average >= 50
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
        
    return average, result

def main():
    print("Student Result Processing System")
    print("-" * 30)
    
    try:
        # Taking multiple marks as input (separated by spaces)
        raw_input = input("Enter student marks separated by spaces (e.g., 60 45 70): ")
        
        # Converting the input string into a list of integers
        marks = [int(m) for m in raw_input.split()]
        
        if len(marks) == 0:
            print("Error: No marks were entered.")
            return

        # Processing the marks
        avg, status = process_student_result(marks)
        
        # Final Output
        print("-" * 30)
        print(f"Total Subjects: {len(marks)}")
        print(f"Average Marks: {avg:.2f}")
        print(f"Final Status:  {status}")
        
    except ValueError:
        print("Error: Please enter only numerical values for marks.")

if __name__ == "__main__":
    main()
```

```python
# Problem 4: Cab Fare Estimator with Retry Option

def calculate_fare(distance, is_peak_hour):
    """
    Computes the total fare based on distance and time of day.
    """
    BASE_FARE = 50
    RATE_PER_KM = 12
    PEAK_SURCHARGE = 1.25 # 25% extra
    
    # Standard calculation: Base + (Distance * Rate)
    subtotal = BASE_FARE + (distance * RATE_PER_KM)
    
    # Apply peak hour multiplier if applicable
    if is_peak_hour:
        final_fare = subtotal * PEAK_SURCHARGE
    else:
        final_fare = subtotal
        
    return final_fare

def main():
    print("Cab Fare Estimation System")
    print("-" * 30)
    
    while True:
        try:
            # Input distance
            dist_input = input("\nEnter distance in km (or type 'exit' to quit): ").lower()
            if dist_input == 'exit':
                print("Thank you for using the Cab Estimator.")
                break
                
            distance = float(dist_input)
            if distance < 0:
                print("Error: Distance cannot be negative.")
                continue

            # Input peak hour status
            peak_input = input("Is it peak hour? (yes/no): ").lower()
            is_peak = peak_input == "yes"

            # Calculate and display results
            total = calculate_fare(distance, is_peak)
            
            print("-" * 20)
            print(f"Distance: {distance} km")
            print(f"Peak Hour Applied: {'Yes' if is_peak else 'No'}")
            print(f"Estimated Fare: ₹{total:.2f}")
            print("-" * 20)

        except ValueError:
            print("Invalid input. Please enter a numerical value for distance.")

if __name__ == "__main__":
    main()
```

```python
# Problem 5: Employee Attendance Eligibility Checker


def check_attendance_eligibility(attendance_list):
    """
    Calculates the percentage of presence and determines eligibility.
    """
    if not attendance_list:
        return 0, "No Data Provided"

    total_days = len(attendance_list)
    present_days = 0

    # Rule 1: Use loop to count present days
    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1
    
    # Rule 2: Calculate percentage
    percentage = (present_days / total_days) * 100
    
    # Rule 3: Return eligibility status (threshold = 75%)
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
        
    return percentage, status

def main():
    print("Employee Attendance Eligibility Checker")
    print("-" * 40)
    
    raw_input = input("Enter attendance (P for Present, A for Absent) separated by spaces: ")
    
    # Processing the raw string into a list
    attendance = raw_input.split()
    
    if not attendance:
        print("Error: No attendance records entered.")
        return

    # Call function and capture returned values
    percent, eligibility = check_attendance_eligibility(attendance)
    
    # Final Output
    print("-" * 40)
    print(f"Total Working Days:  {len(attendance)}")
    print(f"Total Days Present:  {attendance.count('P') + attendance.count('p')}")
    print(f"Attendance Rate:     {percent:.2f}%")
    print(f"Final Status:        {eligibility}")
    print("-" * 40)

if __name__ == "__main__":
    main()
```

```python
# Problem 6: Password Strength Checker


def check_password_strength(password):
    """
    Validates a password based on length, digits, and special characters.
    """
    # Rule 1: Minimum length check
    if len(password) < 8:
        return False, "Password must be at least 8 characters long."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    # Rule 2 & 3: Loop over string to check conditions
    for char in password:
        if char.isdigit():
            has_digit = True
        elif char in special_chars:
            has_special = True

    # Final Verification
    if not has_digit:
        return False, "Password must contain at least one digit (0-9)."
    if not has_special:
        return False, "Password must contain at least one special character (@, #, or $)."

    return True, "Strong Password! Security criteria met."

def main():
    print("Security Portal: Password Strength Checker")
    print("-" * 40)
    
    user_password = input("Enter your password to validate: ")
    
    # Execute validation
    is_strong, message = check_password_strength(user_password)
    
    print("-" * 40)
    if is_strong:
        print(f"Status: VALID")
        print(f"Message: {message}")
    else:
        print(f"Status: INVALID")
        print(f"Reason: {message}")
    print("-" * 40)

if __name__ == "__main__":
    main()
```
