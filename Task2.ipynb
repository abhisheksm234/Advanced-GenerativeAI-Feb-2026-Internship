```python
#Problem 1: Unique Words in a Sentence

def get_unique_words():
    # Take input from the user
    sentence = input("Enter sentence: ")
    
    # Standardize the text by converting to lowercase and splitting into a list
    # .split() automatically handles multiple spaces
    words = sentence.lower().split()
    
    # Using a set to automatically filter out duplicate entries
    unique_words_set = set(words)
    
    # Output the results
    print(f"\nUnique words count: {len(unique_words_set)}")
    print(f"Unique words: {unique_words_set}")

if __name__ == "__main__":
    get_unique_words()
```

```python
#Problem 2: Highest Salary from Employee Data


def find_top_earner():
    # Input data: Employee names and their respective salaries
    employees = {
        "Ravi": 75000,
        "Anita": 68000,
        "Kiran": 72000
    }
    
    # Check if the dictionary is not empty to avoid errors
    if not employees:
        print("No employee data available.")
        return

    # Find the employee name (key) with the maximum salary (value)
    # The 'get' method helps us compare the values associated with the keys
    top_employee = max(employees, key=employees.get)
    highest_salary = employees[top_employee]
    
    # Output the result
    print(f"Highest Salary: {top_employee} - {highest_salary}")

if __name__ == "__main__":
    find_top_earner()
```

```python
#Problem 3: Find Maximum and Minimum Values

def find_extremes(numbers):
    # Check if the list is empty to avoid index errors
    if not numbers:
        return None, None

    # Initialize max and min with the first element of the list
    maximum = numbers[0]
    minimum = numbers[0]

    # Iterate through the list starting from the second element
    for num in numbers:
        # If the current number is greater than our current maximum, update it
        if num > maximum:
            maximum = num
        
        # If the current number is smaller than our current minimum, update it
        elif num < minimum:
            minimum = num
            
    return maximum, minimum

# Input Data
numbers = [45, 22, 89, 10, 66]

# Execution
max_val, min_val = find_extremes(numbers)

# Output formatting
print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

```python
#Problem 4: Count Products Above a Price Threshold


def count_premium_products(prices):
    # Initialize a counter to zero
    count = 0
    threshold = 1000
    
    # Iterate through each price in the provided list
    for price in prices:
        # Check if the current price exceeds our threshold
        if price > threshold:
            count += 1
            
    return count

# Input Data
prices = [450, 1200, 899, 1500, 300]

# Calculate the result
result = count_premium_products(prices)

# Output the result
print(f"Prices: {prices}")
print(f"Products above 1000: {result}")
```

```python
#Problem 5: Calculate Attendance Percentage

def calculate_attendance(attendance_list):
    # Determine the total number of days recorded
    total_days = len(attendance_list)
    
    # Check for an empty list to avoid division by zero
    if total_days == 0:
        return 0.0
    
    # Count how many times "P" (Present) appears in the list
    present_days = attendance_list.count("P")
    
    # Calculate the percentage
    # Formula: (Present Days / Total Days) * 100
    percentage = (present_days / total_days) * 100
    
    return percentage

# Input Data
attendance = ["P", "P", "A", "P", "P"]

# Process the data
result = calculate_attendance(attendance)

# Output formatting
print(f"Attendance: {attendance}")
print(f"Attendance Percentage: {result:.1f}%")
```

```python
#Problem 6: Remove Duplicate Phone Numbers


def clean_contact_list():
    # Input Data: A list containing duplicate phone numbers
    phone_numbers = [9876543210, 9123456789, 9876543210]
    
    # Converting the list to a set to remove duplicates automatically
    # Sets store only unique elements
    unique_contacts = set(phone_numbers)
    
    # Output the result
    print(f"Original list: {phone_numbers}")
    print(f"Unique phone numbers: {unique_contacts}")

if __name__ == "__main__":
    clean_contact_list()
```

```python
#Problem 7: Count Character Frequency

def count_char_frequency():
    # Input string
    text = "pythonp"
    
    # Initialize an empty dictionary to store counts
    frequency = {}
    
    # Iterate through each character in the string
    for char in text:
        # If the character is already in the dictionary, increment its count
        if char in frequency:
            frequency[char] += 1
        # If the character is new, add it to the dictionary with a count of 1
        else:
            frequency[char] = 1
            
    return frequency

# Execute and print results
char_counts = count_char_frequency()
print(f"Character Frequency: {char_counts}")
```

```python
#Problem 8: Convert List to Tuple

def list_to_tuple_conversion():
    # Input Data: A list of integers (Mutable)
    numbers_list = [10, 20, 30]
    
    # Using the built-in tuple() constructor to convert the list
    numbers_tuple = tuple(numbers_list)
    
    # Output the results
    print(f"Original List: {numbers_list}")
    print(f"Converted Tuple: {numbers_tuple}")

if __name__ == "__main__":
    list_to_tuple_conversion()
```

```python
#Problem 9: Check if a Key Exists in a Dictionary


def check_employee_record():
    # Input data: Employee IDs and their roles
    employee_records = {
        "EMP101": "Software Engineer",
        "EMP102": "Data Analyst",
        "EMP103": "Project Manager"
    }

    # Key we want to search for
    search_key = "EMP102"

    # Using the 'in' keyword to check for the key
    if search_key in employee_records:
        print("Employee exists")
        # Bonus: print the value associated with the key
        print(f"Role: {employee_records[search_key]}")
    else:
        print("Employee record not found")

if __name__ == "__main__":
    check_employee_record()
```

```python
#Problem 10: Calculate Average Marks

def calculate_student_average():
    # Input Data: A list of marks obtained by a student
    marks = [85, 70, 65, 70]
    
    # Calculate the total sum of all marks
    total_sum = sum(marks)
    
    # Get the total count of subjects
    num_subjects = len(marks)
    
    # Check for empty list to prevent DivisionByZero error
    if num_subjects == 0:
        return 0.0
    
    # Apply the average formula: Sum / Count
    average = total_sum / num_subjects
    
    # Output the result
    print(f"Marks: {marks}")
    print(f"Average Marks: {average}")

if __name__ == "__main__":
    calculate_student_average()
```
