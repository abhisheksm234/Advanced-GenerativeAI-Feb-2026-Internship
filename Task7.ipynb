```python
# Problem 1: Smart Light Controller


class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False  # Default state is OFF

    def set_status(self, action):
        if action.upper() == "ON":
            self.is_on = True
        elif action.upper() == "OFF":
            self.is_on = False
        else:
            print("Invalid action! Please use ON or OFF.")
            return

    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")

# --- Taking User Input ---
name_input = input("Light Name: ")
action_input = input("Action (ON/OFF): ")

# Initialize the light object
my_light = SmartLight(name_input)

# Update and display status
my_light.set_status(action_input)
my_light.display_status()
```

```python
# Problem 2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\n" + "="*20)
        print(" Employee ID Card ")
        print("="*20)
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")
        print("="*20)

# --- Handling User Input ---
# We collect the raw data from the user first
name_in = input("Employee Name: ")
id_in = input("Employee ID: ")
dept_in = input("Department: ")

# Create an instance of the Employee class
new_employee = Employee(name_in, id_in, dept_in)

# Trigger the display method
new_employee.display_id_card()
```

```python
# Problem 3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# --- Input Section ---
# Collecting data from the user
input_name = input("Contact Name: ")
input_phone = input("Phone Number: ")

# Creating the contact object
new_contact = Contact(input_name, input_phone)

# Displaying the stored information
new_contact.display_contact()
```

```python
# Problem 4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_tag(self):
        # We use an f-string to format the price with the ₹ symbol
        print("\n" + "-" * 15)
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")
        print("-" * 15)

# --- Collecting User Input ---
p_name = input("Product Name: ")
p_price = input("Price: ")

# Creating the product object
# Note: We keep price as a string for simple display, 
# or convert to int/float if calculations are needed.
new_product = Product(p_name, p_price)

# Generating the output
new_product.print_tag()
```

```python
# Problem 5: Movie Rating Display System


class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print("\n" + "*" * 20)
        print(f"Movie: {self.name}")
        # Standardizing the output to show the rating scale
        print(f"Rating: {self.rating} / 5")
        print("*" * 20)

# --- Collecting User Input ---
m_name = input("Movie Name: ")
m_rating = input("Rating: ")

# Creating the movie object
# Even though rating is a number, we can store it as a string for simple display
my_movie = Movie(m_name, m_rating)

# Triggering the display
my_movie.display_details()
```

```python
# Problem 6: Delivery Address Manager


class DeliveryManager:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("\nDelivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# --- User Input Section ---
name_input = input("Customer Name: ")
address_input = input("Address: ")

# Creating the delivery object
order_delivery = DeliveryManager(name_input, address_input)

# Printing the formatted output
order_delivery.display_delivery_details()
```
