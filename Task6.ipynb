```python
# Problem Statement 1:Smart Parking Lot Management System


def manage_parking_lot():
    try:
        # 1. Accept user inputs
        capacity = int(input("Enter Parking Capacity: "))
        logs_input = input("Enter Vehicle Logs (e.g., IN, IN, OUT, IN): ")
        
        # 2. Convert input to a list, handling both commas and spaces
        # We replace commas with spaces and then split to get individual words
        logs = logs_input.replace(',', ' ').split()
        
        current_parked = 0
        peak_usage = 0
        over_capacity_alert = False

        # 3. Loop through logs to calculate totals
        for log in logs:
            action = log.strip().upper()
            if action == "IN":
                current_parked += 1
            elif action == "OUT":
                current_parked = max(0, current_parked - 1)
            
            # Track peak usage
            if current_parked > peak_usage:
                peak_usage = current_parked
            
            # Identify if capacity is exceeded
            if current_parked > capacity:
                over_capacity_alert = True

        # 4. Final status determination
        status = "Full" if current_parked >= capacity else "Available"

        # 5. Output results
        print("\n--- Parking Management Report ---")
        print(f"Currently Parked Vehicles: {current_parked}")
        print(f"Peak Usage Today: {peak_usage}")
        print(f"Parking Status: {status}")

        if over_capacity_alert:
            print("ALERT: Parking capacity was exceeded during this period.")
            
    except ValueError:
        print("Error: Please enter a valid number for capacity.")

# Execute the function
manage_parking_lot()
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator

def estimate_delivery_time():
    try:
        # 1. User Inputs
        distance = float(input("Enter Distance (km): "))
        traffic = input("Enter Traffic Level (Low, Medium, High): ").strip().capitalize()
        weather = input("Enter Weather Condition (Clear, Rainy, Snowy): ").strip().capitalize()

        # 2. Set Base Time (3 minutes per km)
        base_time = distance * 3
        
        # 3. Add Restaurant Prep Time (Constant average)
        prep_time = 15
        
        # 4. Traffic Adjustments
        traffic_delay = 0
        if traffic == "Medium":
            traffic_delay = 10
        elif traffic == "High":
            traffic_delay = 20
            
        # 5. Weather Adjustments
        weather_delay = 0
        if weather == "Rainy":
            weather_delay = 10
        elif weather == "Snowy":
            weather_delay = 20

        # 6. Final Calculation
        total_eta = base_time + prep_time + traffic_delay + weather_delay

        # 7. Output Result
        print("\n--- Delivery Summary ---")
        print(f"Distance: {distance} km")
        print(f"Traffic Delay: {traffic_delay} mins")
        print(f"Weather Delay: {weather_delay} mins")
        print(f"Estimated Delivery Time: {int(total_eta)} minutes")

    except ValueError:
        print("Error: Please enter a numeric value for distance.")

# Run the function
estimate_delivery_time()
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


def analyze_theatre_occupancy():
    try:
        # 1. User Input
        total_seats = int(input("Enter Total Seats: "))
        
        # User can enter 1s separated by spaces or commas
        booked_input = input("Enter Booked Seats (e.g., 1, 1, 1): ")
        
        # 2. Process the list of bookings
        # We clean the input to ensure we only count the '1's or entries
        booked_list = booked_input.replace(',', ' ').split()
        booked_count = len(booked_list)
        
        # 3. Calculate Occupancy Percentage
        if total_seats > 0:
            occupancy_pct = (booked_count / total_seats) * 100
        else:
            occupancy_pct = 0

        # 4. Determine Show Status and Suggestions
        status = "Available"
        suggestion = "Monitor bookings for next hour."
        
        if occupancy_pct >= 100:
            status = "Housefull"
            suggestion = "Urgent: Open additional show immediately."
        elif occupancy_pct >= 80:
            status = "Almost Full"
            suggestion = "Recommended: Schedule an extra show for this movie."
        elif occupancy_pct >= 50:
            status = "Filling Fast"
            suggestion = "No additional show needed yet."

        # 5. Output results
        print("\n--- Theatre Occupancy Report ---")
        print(f"Occupancy: {occupancy_pct:.0f}%")
        print(f"Show Status: {status}")
        print(f"Management Suggestion: {suggestion}")
            
    except ValueError:
        print("Error: Please enter a valid number for total seats.")

# Run the function
analyze_theatre_occupancy()
```

```python
#  Problem Statement 4:Cloud Server Load Classification System

def classify_server_load():
    try:
        # 1. User Input
        # Accepts numbers separated by spaces or commas
        readings_input = input("Enter CPU Readings (%) (e.g., 45, 60, 70): ")
        
        # 2. Clean and process input into a list of floats
        readings = [float(x) for x in readings_input.replace(',', ' ').split()]
        
        if not readings:
            print("Error: No readings provided.")
            return

        # 3. Calculate Average
        avg_load = sum(readings) / len(readings)
        
        # 4. Classification Rules
        status = ""
        action_required = ""
        
        if avg_load < 50:
            status = "Normal"
            action_required = "No action needed. System is healthy."
        elif 50 <= avg_load <= 80:
            status = "Warning"
            action_required = "Monitor closely. Consider scaling resources."
        else:
            status = "Critical"
            action_required = "Immediate Action Required: Potential system failure!"

        # 5. Output Results
        print("\n--- Server Load Analysis Report ---")
        print(f"Number of Data Points: {len(readings)}")
        print(f"Average CPU Load: {avg_load:.0f}%")
        print(f"Server Status: {status}")
        print(f"Recommendation: {action_required}")

    except ValueError:
        print("Error: Please enter valid numerical percentages.")

# Run the function
classify_server_load()
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor


def monitor_classroom_resources():
    # 1. Define thresholds (Maximum recommended hours per day)
    thresholds = {
        "Projector": 5,
        "AC": 8,
        "Lights": 10
    }

    # 2. User Input
    print("Enter daily usage hours for the following resources:")
    try:
        projector_hrs = float(input("Projector: "))
        ac_hrs = float(input("AC: "))
        lights_hrs = float(input("Lights: "))

        # 3. Store in a dictionary
        usage = {
            "Projector": projector_hrs,
            "AC": ac_hrs,
            "Lights": lights_hrs
        }

        # 4. Identify Overused Resources
        overused = []
        energy_alert = "No"

        for resource, hours in usage.items():
            # Check against specific threshold for that resource
            if hours > thresholds[resource]:
                overused.append(resource)
                energy_alert = "Yes"

        # 5. Output Results
        print("\n--- Resource Usage Report ---")
        if overused:
            print(f"Overused Resources: {', '.join(overused)}")
        else:
            print("Overused Resources: None")
            
        print(f"Energy Alert: {energy_alert}")

    except ValueError:
        print("Error: Please enter a valid number for usage hours.")

# Run the function
monitor_classroom_resources()
```

```python
#  Problem Statement 6: Online Event Registration Capacity Controller


def manage_event_registrations():
    try:
        # 1. User Input
        capacity = int(input("Enter Event Capacity: "))
        total_registrations = int(input("Enter Total Registrations: "))

        # 2. Logic to calculate confirmed and waitlisted users
        if total_registrations > capacity:
            confirmed = capacity
            waitlist = total_registrations - capacity
            status = "Closed"
        else:
            confirmed = total_registrations
            waitlist = 0
            status = "Open"

        # 3. Output Results
        print("\n--- Event Registration Report ---")
        print(f"Confirmed Registrations: {confirmed}")
        print(f"Waitlisted Users: {waitlist}")
        print(f"Registration Status: {status}")

    except ValueError:
        print("Error: Please enter valid whole numbers for capacity and registrations.")

# Run the function
manage_event_registrations()
```
