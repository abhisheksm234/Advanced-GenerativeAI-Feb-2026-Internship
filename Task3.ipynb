```python
# Problem Statement 1: Employee Performance Bonus Eligibility


# 1. Take input from the user
# We'll expect a format like "Name: Score, Name: Score"
print("Enter employees and scores (e.g., Ravi: 92, Anita: 88):")
raw_input = input("> ")

# 2. Parse the input into a dictionary
employees = {}
for entry in raw_input.split(","):
    name, score = entry.split(":")
    employees[name.strip()] = int(score.strip())

# 3. Identify the highest performance score
if not employees:
    print("No employee data provided.")
else:
    max_score = max(employees.values())

    # 4. Handle ties by finding all employees with that score
    top_performers = [name for name, score in employees.items() if score == max_score]

    # 5. Display the output
    performers_list = ", ".join(top_performers)
    print(f"\nTop Performers Eligible for Bonus: {performers_list} (Score: {max_score})")
```

```python
# Problem Statement 2: Search Query Keyword Analysis


import string

# 1. Take input from the user
user_input = input("Enter your search query: ")

# 2. Preprocessing
# Convert to lowercase
text = user_input.lower()

# Remove common punctuation using the string.punctuation constant
# This replaces punctuation marks with an empty string
text = text.translate(str.maketrans('', '', string.punctuation))

# 3. Tokenize (Split into words)
words = text.split()

# 4. Count frequencies
word_counts = {}
for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

# 5. Filter for keywords searched more than once
repeated_keywords = {word: count for word, count in word_counts.items() if count > 1}

# 6. Display output
print(f"\nRepeated Keywords: {repeated_keywords}")
```

```python
# Problem Statement 3: Sensor Data Validation

# 1. Take input from the user
# We expect a comma-separated string like: 3, 4, 7, 8, 10, 12, 5
raw_input = input("Enter sensor readings separated by commas: ")

# 2. Convert the input string into a list of integers
try:
    sensor_readings = [int(val.strip()) for val in raw_input.split(",")]
    
    # 3. Use enumerate to track the hour (index) and the value
    # We filter using a list comprehension: if reading % 2 == 0
    valid_readings = [
        (hour, reading) 
        for hour, reading in enumerate(sensor_readings) 
        if reading % 2 == 0
    ]

    # 4. Display the output
    print("\nValid Sensor Readings (Hour, Value):")
    print(valid_readings)

except ValueError:
    print("Error: Please ensure you only enter numbers separated by commas.")
```

```python
# Problem Statement 4: Email Domain Usage Analysis


# 1. Take input from the user
print("Enter employee email IDs separated by commas:")
raw_input = input("> ")

# 2. Clean and parse the input into a list
emails = [email.strip() for email in raw_input.split(",") if "@" in email]

if not emails:
    print("No valid emails provided.")
else:
    # 3. Extract domains and count frequencies
    domain_counts = {}
    for email in emails:
        # Split by '@' and take the second part
        domain = email.split("@")[1]
        domain_counts[domain] = domain_counts.get(domain, 0) + 1

    # 4. Calculate total and percentages
    total_emails = len(emails)
    
    print("\nEmail Domain Usage Analysis:")
    for domain, count in domain_counts.items():
        # Calculate percentage: (Part / Total) * 100
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")
```

```python
# Problem Statement 5: Sales Spike Detection

# 1. Take input from the user
print("Enter daily sales values separated by commas (e.g., 1200, 1500, 900):")
raw_input = input("> ")

try:
    # 2. Convert input string into a list of integers
    sales = [int(s.strip()) for s in raw_input.split(",")]
    
    if not sales:
        print("No sales data provided.")
    else:
        # 3. Calculate the average daily sales
        average_sales = sum(sales) / len(sales)
        
        # 4. Define the spike threshold (30% above average)
        # Threshold = Average * 1.30
        threshold = average_sales * 1.30
        
        print(f"\nAverage Sales: {average_sales:.2f}")
        print(f"Spike Threshold (Avg + 30%): {threshold:.2f}")
        print("-" * 30)
        
        # 5. Detect and display spikes
        # We use enumerate(sales, 1) to start day numbering from 1 instead of 0
        found_spike = False
        for day, value in enumerate(sales, 1):
            if value > threshold:
                print(f"Day {day}: {value}")
                found_spike = True
        
        if not found_spike:
            print("No significant sales spikes detected.")

except ValueError:
    print("Error: Please enter valid numbers separated by commas.")
```

```python
# Problem Statement 6: Duplicate User ID Detection

# 1. Take input from the user
print("Enter User IDs separated by commas (e.g., user1, user2, user1):")
raw_input = input("> ")

# 2. Parse the input into a list and clean whitespace
user_ids = [uid.strip() for uid in raw_input.split(",") if uid.strip()]

if not user_ids:
    print("No User IDs provided.")
else:
    # 3. Create a frequency map (dictionary) of the IDs
    id_counts = {}
    for uid in user_ids:
        id_counts[uid] = id_counts.get(uid, 0) + 1
    
    # 4. Identify and display only the duplicates (count > 1)
    print("\nDuplicate User ID Report:")
    has_duplicates = False
    
    for uid, count in id_counts.items():
        if count > 1:
            print(f"{uid} → {count} times")
            has_duplicates = True
            
    if not has_duplicates:
        print("No duplicate User IDs detected. All IDs are unique.")
```
