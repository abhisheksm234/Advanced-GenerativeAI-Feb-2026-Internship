```python
# Problem 1: Social Media – Post Engagement Analyze

def analyze_user_input():
    print("--- Social Media Engagement Tracker ---")
    
    # Prompt the user for a list of like counts separated by spaces
    raw_input = input("Enter like counts (e.g., 200 450 300 100): ")
    
    # Convert the string input into a list of integers
    # .split() breaks the string at spaces; int(x) converts each piece to a number
    likes_list = [int(x) for x in raw_input.split()]
    
    total_likes = 0
    for likes in likes_list:
        total_likes += likes
    
    # Logic check
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Final Output
    print("-" * 30)
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Run the analyzer
analyze_user_input()
```

```python
# Problem 2: Healthcare – Medicine Stock Alert System

def check_medicine_stock():
    # Take user input for the medicine name and the current count
    medicine_name = input("Enter Medicine Name: ")
    try:
        current_stock = int(input(f"Enter current stock for {medicine_name}: "))
        
        # Logic to determine status based on the threshold of 10
        if current_stock < 10:
            status = "Low Stock Alert"
        else:
            status = "Stock Sufficient"
            
        # Display the results in the requested format
        print("\n--- Inventory Report ---")
        print(f"Medicine: {medicine_name}")
        print(f"Medicine Stock: {current_stock}")
        print(f"Status: {status}")
        
    except ValueError:
        print("Invalid input! Please enter a whole number for the stock quantity.")

# Execute the function
check_medicine_stock()
```

```python
# Problem 3: Agriculture – Rainfall Adequacy Checker


def check_rainfall_adequacy():
    print("--- Seasonal Rainfall Analyzer ---")
    
    # User inputs: Required threshold and actual readings
    try:
        required_level = float(input("Enter required average rainfall (e.g., 50): "))
        raw_data = input("Enter rainfall readings separated by spaces: ")
        
        # Convert string input into a list of floats
        readings = [float(x) for x in raw_data.split()]
        
        if not readings:
            print("No data entered.")
            return

        # Calculate average using a loop
        total_rainfall = 0
        for value in readings:
            total_rainfall += value
        
        average_rainfall = total_rainfall / len(readings)
        
        # Determine status
        if average_rainfall >= required_level:
            status = "Adequate Rainfall"
        else:
            status = "Inadequate Rainfall"
            
        # Display Output
        print("-" * 30)
        print(f"Average Rainfall: {average_rainfall:.2f}")
        print(f"Rainfall Status: {status}")
        
    except ValueError:
        print("Error: Please enter numeric values only.")

# Run the program
check_rainfall_adequacy()
```

```python
# Problem 4: Social Media – Duplicate Account Detection

def detect_duplicates():
    print("--- Account Uniqueness Verifier ---")
    
    # Prompt user for a list of usernames
    raw_input = input("Enter usernames separated by spaces: ")
    
    # Create a list of usernames from the input
    username_list = raw_input.lower().split() # Lowercase to catch "User1" and "user1"
    
    if not username_list:
        print("No usernames entered.")
        return

    # Use a set to find unique usernames
    unique_usernames = set(username_list)
    
    # If the length of the set is less than the list, a duplicate existed
    if len(unique_usernames) < len(username_list):
        result = "Yes"
    else:
        result = "No"
    
    # Display Output
    print("-" * 30)
    print(f"Duplicate Accounts Found: {result}")

# Execute the detector
detect_duplicates()
```

```python
# Problem 5: Healthcare – Appointment Eligibility Checker


def check_appointment_eligibility():
    print("--- Healthcare Appointment Portal ---")
    
    try:
        # Take patient age as input
        age_input = input("Enter Patient Age: ")
        patient_age = int(age_input)
        
        # Core Logic: Threshold check for adulthood/legal consent
        if patient_age >= 18:
            status = "Eligible"
        else:
            status = "Not Eligible"
            
        # Display Output in the requested format
        print("-" * 30)
        print(f"Patient Age: {patient_age}")
        print(f"Eligibility Status: {status}")
        
    except ValueError:
        print("Error: Please enter a valid number for age.")

# Run the function
check_appointment_eligibility()
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

def filter_premium_crops():
    print("--- Agri-Market Price Analyzer ---")
    
    try:
        # Prompt user for a list of crop prices
        raw_input = input("Enter crop prices (e.g., 1500 2500 1800 3200): ")
        
        # Convert the string input into a list of integers
        price_list = [int(p) for p in raw_input.split()]
        
        if not price_list:
            print("No prices entered.")
            return

        # Filtering logic: Only keep prices > 2000
        premium_crops = [price for price in price_list if price > 2000]
        
        # Display the result in the requested format
        print("-" * 30)
        print(f"Premium Crops: {premium_crops}")
        
    except ValueError:
        print("Error: Please enter numbers only (e.g., 1200 2100).")

# Execute the filter
filter_premium_crops()
```

```python
# Problem 7: System Monitoring – Application Health Checker

def check_app_health():
    print("--- System Health Operations Dashboard ---")
    
    try:
        # Prompt user for the current error count
        error_input = input("Enter current error count: ")
        error_count = int(error_input)
        
        # Core Logic: Categorizing system health based on error thresholds
        if error_count == 0:
            status = "Healthy"
        elif error_count <= 5:
            status = "Minor Issues"
        else:
            status = "Critical Issues"
            
        # Display Output in the requested format
        print("-" * 35)
        print(f"Error Count: {error_count}")
        print(f"System Status: {status}")
        
    except ValueError:
        print("Error: Please enter a valid integer for the error count.")

# Execute the health check
check_app_health()
```

```python
##### Problem 8: Banking – Daily Transaction Limit Checker


def check_transaction_limit():
    # Define the fixed daily limit
    DAILY_LIMIT = 50000
    
    print("--- Secure Banking Gateway ---")
    
    try:
        # Prompt user for the transaction amount
        user_input = input("Enter Transaction Amount (₹): ")
        transaction_amount = float(user_input)
        
        # Validation Logic
        if transaction_amount <= DAILY_LIMIT:
            status = "Approved"
        else:
            status = "Rejected"
            
        # Display Output in the requested format
        print("-" * 30)
        print(f"Transaction Amount: {transaction_amount}")
        print(f"Transaction Status: {status}")
        
    except ValueError:
        print("Error: Please enter a valid numerical amount.")

# Run the checker
check_transaction_limit()
```

```python
# Problem 9: E-Learning – Student Attendance Eligibility System


def check_exam_eligibility():
    print("--- Academic Attendance Portal ---")
    
    try:
        # Prompt user for attendance data (1 for Present, 0 for Absent)
        raw_input = input("Enter attendance records (e.g., 1 1 0 1 0): ")
        
        # Convert string input into a list of integers
        records = [int(x) for x in raw_input.split()]
        
        if not records:
            print("No attendance records provided.")
            return

        # Use a loop to calculate total days present
        days_present = 0
        for day in records:
            if day == 1:
                days_present += 1
        
        # Calculate Percentage
        total_days = len(records)
        attendance_percentage = (days_present / total_days) * 100
        
        # Core Logic: Threshold check (75%)
        if attendance_percentage >= 75:
            eligibility = "Eligible"
        else:
            eligibility = "Not Eligible"
            
        # Display Output in the requested format
        print("-" * 35)
        print(f"Attendance Percentage: {attendance_percentage:.1f}")
        print(f"Exam Eligibility: {eligibility}")
        
    except ValueError:
        print("Error: Please enter only 1s and 0s (space-separated).")

# Execute the checker
check_exam_eligibility()
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill():
    print("--- Smart Electricity Billing System ---")
    
    try:
        # Input: Total units consumed
        units = float(input("Enter units consumed: "))
        
        if units < 0:
            print("Units cannot be negative.")
            return

        # 1. Calculate Total Bill using slabs
        total_bill = 0
        
        if units <= 100:
            total_bill = units * 3
        elif units <= 200:
            # First 100 units at ₹3, remaining at ₹5
            total_bill = (100 * 3) + ((units - 100) * 5)
        else:
            # First 100 at ₹3, second 100 at ₹5, remainder at ₹7
            total_bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
            
        # 2. Classify Usage Level
        if total_bill < 500:
            usage_status = "Low Usage"
        elif 500 <= total_bill <= 1500:
            usage_status = "Moderate Usage"
        else:
            usage_status = "High Usage"
            
        # Display the output
        print("-" * 35)
        print(f"Total Bill: ₹{total_bill:.2f}")
        print(f"Usage Status: {usage_status}")
        
    except ValueError:
        print("Error: Please enter a valid numerical value for units.")

# Execute the analyzer
calculate_electricity_bill()
```
