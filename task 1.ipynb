```python
def analyze_results():
    # 1. Take input from the user
    print("--- Academic Evaluation System ---")
    raw_input = input("Enter student marks separated by spaces (e.g., 45 78 90): ")
    
    # Convert the input string into a list of integers
    try:
        marks = [int(m) for m in raw_input.split()]
    except ValueError:
        print("Error: Please enter numbers only.")
        return

    # 2. Initialize counters
    pass_count = 0
    fail_count = 0
    
    # 3. Analyze marks
    for m in marks:
        if m >= 50:
            pass_count += 1
        else:
            fail_count += 1
            
    # 4. Print the final result clearly
    print("\n--- Final Analysis ---")
    print(f"Total Students Evaluated: {len(marks)}")
    print(f" Pass Count: {pass_count}")
    print(f" Fail Count: {fail_count}")
    
    # Simple success rate calculation
    if len(marks) > 0:
        pass_percentage = (pass_count / len(marks)) * 100
        print(f"Overall Pass Percentage: {pass_percentage:.1f}%")

# Run the function
analyze_results()
```

```python
def clean_user_data():
    # 1. Take input from the user
    print("--- Data Preprocessing Tool ---")
    raw_input = input("Enter names separated by commas (e.g.,  Alice , bob, CHARLIE ): ")
    
    # Split the string into a list using the comma as a delimiter
    raw_names = raw_input.split(",")
    
    # 2. Clean and standardize
    # We use a list comprehension for a clean, one-line transformation
    cleaned_names = [name.strip().lower() for name in raw_names]
    
    # 3. Print the final result
    print("\n--- Cleaned Results ---")
    print(f"Original Count: {len(raw_names)}")
    print(f"Cleaned List: {cleaned_names}")

# Run the cleaner
clean_user_data()
```

```python
def analyze_messages():
    # 1. Take input from the user
    print("--- Message Length Analyzer ---")
    raw_input = input("Enter messages separated by commas: ")
    
    # Split the input into a list of messages
    messages = [msg.strip() for msg in raw_input.split(",")]
    
    print("\n--- Analysis Results ---")
    
    # 2. Iterate and Flag
    for msg in messages:
        length = len(msg)
        
        # Check if the message exceeds 10 characters
        if length > 10:
            status = "[LONG]"
        else:
            status = "[SHORT]"
            
        print(f"{status} | Length: {length} | Message: \"{msg}\"")

# Run the analyzer
analyze_messages()
```

```python
def analyze_logs():
    # 1. Take input from the user
    print("--- System Log Monitor ---")
    raw_input = input("Enter log levels separated by spaces (e.g., INFO ERROR WARNING ERROR): ")
    
    # Standardize input to uppercase to ensure the search is case-insensitive
    logs = [log.strip().upper() for log in raw_input.split()]
    
    # 2. Count the "ERROR" entries
    # Method: Using the built-in .count() list method for efficiency
    error_count = logs.count("ERROR")
    total_logs = len(logs)
    
    # 3. Print the final result clearly
    print("\n--- Log Analysis Report ---")
    print(f"Total Logs Processed: {total_logs}")
    
    if error_count > 0:
        print(f" Total ERROR Count: {error_count}")
        # Calculate error density
        error_rate = (error_count / total_logs) * 100
        print(f"System Health: {100 - error_rate:.1f}% (Error Rate: {error_rate:.1f}%)")
    else:
        print(" No errors detected. System is healthy.")

# Run the detector
analyze_logs()
```
